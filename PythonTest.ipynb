Dataset Information: We have 3 files:
- car_data: This table has information about cars and their retail price
- car_info: Has additional information abour the cars
- car_used_prices: This  table has the Used prices of the cars

```python
import pandas as pd
import os
import numpy as np

# loading excel
car_data = pd.read_csv("yt_car_main.csv")
car_info = pd.read_csv("yt_car_info.csv")
car_used_prices = pd.read_csv("yt_car_used_prices.csv")

# to see your directory: os.getcwd()
pd.set_option('display.max_rows', None) # display all rows
pd.set_option('display.max_columns', None) # display all columns

car_data.head()
```

### 1. What is the average retail_price per model for 2023. Order desc

```python
car_data_2023 = car_data[car_data['Year'] == 2023]

car_data_2023[['Model', 'Retail_Price']].groupby('Model').agg({'Retail_Price' : 'mean'}).sort_values(by='Retail_Price', ascending=False)
```

### 2. What is the total retail_price per make and Cylinders for AWD cars. Order by desc

```python
awd_cars = pd.merge(car_data, car_info, on='index')
awd_cars = awd_cars[awd_cars['Drivetrain'] == 'AWD']

awd_cars[['Model', 'Cylinders', 'Retail_Price']].groupby(['Model', 'Cylinders']).agg({'Retail_Price' : 'sum'}).sort_values(by='Retail_Price', ascending=False)
```

### 3. Show the average Used Prices per Model and year for cars that have 500+ horsepower

```python
used_cars = pd.merge(car_data, car_used_prices, on='index')

used_cars = pd.merge(used_cars, car_info, on='index')

used_cars['Horsepower'] = used_cars['Horsepower'].str[:3].astype(float)

used_cars = used_cars[used_cars['Horsepower'] >= 500]

used_cars[['Year', 'Model', 'Used Price']].groupby(['Model', 'Year']).agg({'Used Price' : 'mean'})
```

### 4. Show the top 10 models and make in terms of highway fuel economy (min) that have a retail price less than 50k

```python
top_10 = car_data[car_data['Retail_Price'] < 50000]

top_10 = pd.merge(top_10, car_info, on='index')

top_10['Highway Fuel Economy'].str.replace(' mpg', '').astype(float)

top_10[['Model', 'Make','Highway Fuel Economy']].sort_values(by='Highway Fuel Economy', ascending=False).head(10)
```
